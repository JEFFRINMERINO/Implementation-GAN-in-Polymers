```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

DATA_PATH = r"D:\Bio_Materials\antibiofilm_polymers.csv"
df = pd.read_csv(DATA_PATH)

print("Shape:", df.shape)
print("\nMissing values:")
print(df.isnull().sum())

print("\nTarget statistics:")
print(df["antibiofilm"].describe())
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.drop(columns=["polymer_id"]).corr(),
            annot=True,
            cmap="coolwarm",
            fmt=".2f")
plt.title("Feature Correlation Matrix")
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

features = [
    "hydrophobicity",
    "surface_charge",
    "surface_roughness",
    "hbd",
    "hba",
    "elastic_modulus"
]

X = df[features].values
y = df["antibiofilm"].values

poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X)

print("Original feature count:", X.shape[1])
print("Expanded feature count:", X_poly.shape[1])
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import RidgeCV
from sklearn.model_selection import train_test_split

# Scale
scaler = StandardScaler()
X_poly_scaled = scaler.fit_transform(X_poly)

# Split
X_train, X_test, y_train, y_test = train_test_split(
    X_poly_scaled,
    y,
    test_size=0.2,
    random_state=42
)

# Ridge with automatic alpha selection
alphas = np.logspace(-3, 3, 50)
ridge = RidgeCV(alphas=alphas, cv=5)

ridge.fit(X_train, y_train)

print("Best alpha:", ridge.alpha_)
```

```python
from sklearn.metrics import r2_score, mean_squared_error

y_train_pred = ridge.predict(X_train)
y_test_pred = ridge.predict(X_test)

print("Train R2:", r2_score(y_train, y_train_pred))
print("Test R2:", r2_score(y_test, y_test_pred))
print("Test MSE:", mean_squared_error(y_test, y_test_pred))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

residuals = y_test - y_test_pred

plt.figure(figsize=(7,5))
sns.scatterplot(x=y_test_pred, y=residuals, alpha=0.5)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel("Predicted Score")
plt.ylabel("Residual")
plt.title("Residual Plot")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.regplot(x=y_test, y=y_test_pred,
            scatter_kws={"alpha":0.4},
            line_kws={"color":"red"})
plt.xlabel("True Score")
plt.ylabel("Predicted Score")
plt.title("True vs Predicted (Ridge Model)")
plt.show()
```

```python
import torch
from torch.utils.data import DataLoader, TensorDataset

# Use original 6 scaled features (not polynomial)
scaler_gan = StandardScaler()
X_gan_scaled = scaler_gan.fit_transform(X)

X_tensor = torch.tensor(X_gan_scaled, dtype=torch.float32)

dataset = TensorDataset(X_tensor)
loader = DataLoader(dataset, batch_size=128, shuffle=True)

print("GAN training data shape:", X_tensor.shape)
```

```python
import torch.nn as nn

latent_dim = 12
feature_dim = 6

class Generator(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = nn.Sequential(
            nn.Linear(latent_dim, 64),
            nn.BatchNorm1d(64),
            nn.LeakyReLU(0.2),
            
            nn.Linear(64, 128),
            nn.BatchNorm1d(128),
            nn.LeakyReLU(0.2),
            
            nn.Linear(128, feature_dim),
            nn.Tanh()   # Output bounded between -1 and 1
        )
        
    def forward(self, z):
        return self.model(z)

class Discriminator(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = nn.Sequential(
            nn.Linear(feature_dim, 128),
            nn.LeakyReLU(0.2),
            
            nn.Linear(128, 64),
            nn.LeakyReLU(0.2),
            
            nn.Linear(64, 1),
            nn.Sigmoid()
        )
        
    def forward(self, x):
        return self.model(x)

G = Generator()
D = Discriminator()

print("Improved GAN initialized.")
```

```python
import torch

criterion = nn.BCELoss()

optimizer_G = torch.optim.Adam(G.parameters(), lr=0.0003, betas=(0.5, 0.999))
optimizer_D = torch.optim.Adam(D.parameters(), lr=0.0003, betas=(0.5, 0.999))

epochs = 400

losses_G = []
losses_D = []

for epoch in range(epochs):
    for real_batch in loader:
        
        real_data = real_batch[0]
        batch_size = real_data.size(0)
        
        # ---- Label smoothing ----
        real_labels = torch.full((batch_size, 1), 0.9)
        fake_labels = torch.zeros(batch_size, 1)
        
        # =========================
        # Train Discriminator
        # =========================
        z = torch.randn(batch_size, latent_dim)
        fake_data = G(z)
        
        D_real = D(real_data)
        D_fake = D(fake_data.detach())
        
        loss_real = criterion(D_real, real_labels)
        loss_fake = criterion(D_fake, fake_labels)
        loss_D = loss_real + loss_fake
        
        optimizer_D.zero_grad()
        loss_D.backward()
        
        torch.nn.utils.clip_grad_norm_(D.parameters(), 1.0)
        optimizer_D.step()
        
        # =========================
        # Train Generator
        # =========================
        z = torch.randn(batch_size, latent_dim)
        fake_data = G(z)
        D_fake = D(fake_data)
        
        loss_G = criterion(D_fake, real_labels)
        
        optimizer_G.zero_grad()
        loss_G.backward()
        
        torch.nn.utils.clip_grad_norm_(G.parameters(), 1.0)
        optimizer_G.step()
    
    losses_G.append(loss_G.item())
    losses_D.append(loss_D.item())
    
    if (epoch + 1) % 50 == 0:
        print(f"Epoch [{epoch+1}/{epochs}] | D Loss: {loss_D.item():.4f} | G Loss: {loss_G.item():.4f}")

print("Improved GAN training complete.")
```

```python
# Generate synthetic samples
z = torch.randn(3000, latent_dim)
synthetic_scaled = G(z).detach().numpy()

# Convert back to original scale
synthetic_unscaled = scaler_gan.inverse_transform(synthetic_scaled)

print("Synthetic shape:", synthetic_unscaled.shape)
print("\nFirst 5 synthetic polymers:")
print(synthetic_unscaled[:5])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

feature_names = features

for i, name in enumerate(feature_names):
    plt.figure(figsize=(6,4))
    sns.kdeplot(X[:, i], label="Real", fill=True)
    sns.kdeplot(synthetic_unscaled[:, i], label="Synthetic", fill=True)
    plt.title(f"Distribution Comparison: {name}")
    plt.legend()
    plt.show()
```

```python
from sklearn.decomposition import PCA

# Combine for joint PCA
combined = np.vstack([X_gan_scaled, synthetic_scaled])

pca = PCA(n_components=2)
pca_result = pca.fit_transform(combined)

real_pca = pca_result[:len(X_gan_scaled)]
fake_pca = pca_result[len(X_gan_scaled):]

plt.figure(figsize=(6,5))
plt.scatter(real_pca[:,0], real_pca[:,1], alpha=0.3, label="Real")
plt.scatter(fake_pca[:,0], fake_pca[:,1], alpha=0.3, label="Synthetic")
plt.legend()
plt.title("PCA: Real vs Synthetic Distribution")
plt.show()
```

```python
# Transform synthetic features to polynomial space
synthetic_poly = poly.transform(synthetic_unscaled)

# Scale using same scaler used for ridge
synthetic_poly_scaled = scaler.transform(synthetic_poly)

# Predict using trained Ridge
synthetic_scores = ridge.predict(synthetic_poly_scaled)

print("Synthetic score stats:")
print("Min:", np.min(synthetic_scores))
print("Max:", np.max(synthetic_scores))
print("Mean:", np.mean(synthetic_scores))
```

```python
top_k = 10

top_indices = np.argsort(synthetic_scores)[-top_k:]
top_features = synthetic_unscaled[top_indices]
top_scores = synthetic_scores[top_indices]

print("Top 10 predicted antibiofilm scores:")
print(top_scores)

print("\nTop 10 generated polymer features:")
print(top_features)
```
